```python
from bs4 import BeautifulSoup
import requests
import csv

url = 'https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&sort=relevance'
req = requests.get(url)
website = BeautifulSoup(req.content, 'lxml')
```

```python
print(soup.prettify())
```

```python
website = soup.find('div', class_='_1YokD2 _3Mn1Gg')
```

```python
print(website.prettify())
```

```python
for i in website.find_all('div', class_='_1AtVbE col-12-12'):
    try:
        link = i.find('a', class_='_1fQZEK')['href']
        print('www.flipkart.com' +  link)
    except Exception as e:
        link = None
                
    try:
        image_link = i.find('img', class_='_396cs4')['src']
        print('Link: ',link)
    except Exception as e:
        image_link = None
        
    try:
        name = i.find('div',class_='_4rR01T').text.split('(',1)[0]
        print('Name: ',name)
    except Exception as e:
        name = None
    try:
        color = i.find('div',class_='_4rR01T').text.split('(',1)[1]
        print('Color: ',color.split(',',1)[0])
    except:
        color = None
        
    try:
        ram = i.find('li',class_='rgWa7D').text.split('|',1)[0]
        rom = i.find('li',class_='rgWa7D').text.split('|',1)[1]
        print('RAM: ', ram.replace('RAM',''))
        print('ROM: ', rom.replace('ROM',''))
    except:
        ram = None
        rom = None
    try:
        google_rating = i.find('span', class_='_1lRcqv').text
        print('Rating: ',google_rating)
    except Exception as e:
        google_rating = None
    
    try:
        public_ratings = i.find('span', class_='_2_R_DZ').text.split('&',1)[0]
        print('Ratings: ',public_ratings.replace('Ratings',''))
    except Exception as e:
        public_ratings = None
        
    try:
        reviews = i.find('span', class_='_2_R_DZ').text.split('&',1)[1]
        print('Reviews: ',reviews.replace('Reviews',''))
    except Exception as e:
        reviews = None
    
    try:
        list = i.find_all('li', class_='rgWa7D')
        for j in list:
            configuration = j.text + (",")
            print(configuration)
        # print()
    except Exception as e:
        configuration = None
        
    try:
        ron = i.find_all('li', class_='rgWa7D')
        for i in ron:
            fon = i.text
            print('Fon: ',fon)
            i+1
            ton = i.text
            print('TOn: ', ton)
            i+1
    except Exception as e:
        ron = None
        
    try:
        price = i.find('div', class_='_30jeq3 _1_WHN1').text
        print('Price: ',price)
        print('-------------------------------------------------------------------------------')
    except Exception as e:
        price = None
```

```python
csv_file = open('mobile_data.csv','w',  encoding="utf-8")

csv_write = csv.writer(csv_file)
csv_write.writerow(['name','RAM','ROM','google_rating','public_rating','reviews','configuration','price'])

# site = ['relevance','price_desc']
# for j in site:
for k in range(1,41):
    webpage = requests.get('https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&sort=relevance&page={}'.format(k)).text
    # req = requests.get(url)
    soup = BeautifulSoup(webpage, 'lxml')
    website = soup.find('div', class_='_1YokD2 _3Mn1Gg')
    for i in website.find_all('div', class_='_1AtVbE col-12-12'):
    
        try:
            product = i.find('div',class_='_4rR01T').text.split('(',1)[0]
            # print('Name: ',name)
            name = product
        except:
            name = None
        try:
            ram = i.find('li',class_='rgWa7D').text.split('|',1)[0]
            rom = i.find('li',class_='rgWa7D').text.split('|',1)[1]
            # print('RAM: ', ram.replace('RAM',''))
            # print('ROM: ', rom.replace('ROM',''))
        except:
            ram = None
            rom = None
            
        try:
            rate = i.find('span', class_='_1lRcqv').text
            # print('Rating: ',rating)
            rating = rate
        except:
            rating = None

        try:
            rates = i.find('span', class_='_2_R_DZ').text.split('&',1)[0]
            # print('Ratings: ',ratings)
            ratings = rates.replace('Ratings','')
        except:
            ratings = None

        try:
            review = i.find('span', class_='_2_R_DZ').text.split('&',1)[1]
            # print('Reviews: ',reviews)
            reviews = review.replace('Reviews','')
        except:
            reviews = None
            # configuration = i.find('li',class_='rgWa7D').text
            # print('\n',configuration)

        try:
            list = i.find_all('ul', class_='_1xgFaf')
            for j in list:
                # con = j.text
                # con = j.text + (',')
                configuration = j.text + (',')
            # print()
        except:
            configuration = None

        try:
            amount = i.find('div', class_='_30jeq3 _1_WHN1').text
            # print('Price: ',price)
            price = amount
            # print('-------------------------------------------------------------------------------')
        except:
            price = None
                
        csv_write.writerow([name, ram, rom, rating, ratings, reviews, configuration, price])

csv_file.close()
```
